```python
from peft import LoraConfig, get_peft_model
from unsloth.chat_templates import get_chat_template
from unsloth import FastLanguageModel, is_bfloat16_supported
from trl import SFTTrainer
from transformers import TrainingArguments, TextStreamer
```



1. **Max Sequence Length**: Set to **5020 tokens** to allow the model to process longer texts, enhancing context understanding as needed.

2. **Loading LLAMA 3.1 Model**: Utilizes `FastLanguageModel.from_pretrained` to load the model **"unsloth/Meta-Llama-3.1-8B-bnb-4bit"** with **4-bit precision** (`load_in_4bit=True`), improving memory efficiency and processing speed.

3. **Applying PEFT (Parameter-Efficient Fine-Tuning)**: Configures **LoRA** techniques for fine-tuning specific layers, reducing resource demands. Key parameters include `r=16`, `lora_alpha=16`, and targeted attention layers. Enabling `use_rslora=True` and `use_gradient_checkpointing="unsloth"` enhances stability and memory efficiency.

4. **Verifying Trainable Parameters**: Calls `model.print_trainable_parameters()` to ensure that only selected components of the model are updated during fine-tuning.

```python
max_seq_length = 5020
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Meta-Llama-3.1-8B-bnb-4bit",
    max_seq_length=max_seq_length,
    load_in_4bit=True,
    dtype=None,
)

model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=16,
    lora_dropout=0,
    target_modules=["q_proj", "k_proj", "v_proj", "up_proj", "down_proj", "o_proj", "gate_proj"],
    use_rslora=True,
    use_gradient_checkpointing="unsloth"
)
print(model.print_trainable_parameters())
```

# Prapare data for model feed
Design a prompt format for mental health analysis that examines text for indicators of emotional distress, coping mechanisms, and overall mental well-being. Highlight concerns or positives with brief explanations for each observation, ensuring clear input-output formatting for model processing.








```python
data_prompt = """Analyze the provided text from a mental health perspective. Identify any indicators of emotional distress, coping mechanisms, or psychological well-being. Highlight any potential concerns or positive aspects related to mental health, and provide a brief explanation for each observation.

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token
def formatting_prompt(examples):
    inputs = examples["Context"]
    outputs = examples["Response"]
    texts = []
    for input_, output in zip(inputs, outputs):
        text = data_prompt.format(input_, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
```

**Format the data for training**

```python
from datasets import Dataset
```

```python
training_data = Dataset.from_pandas(filtered_data)
training_data = training_data.map(formatting_prompt, batched=True)
```


# Training setup to start fine tuning!



**Trainer Initialization:**

We initialize the SFTTrainer with the model and tokenizer, along with the training dataset (training_data). The dataset_text_field parameter specifies the field in the dataset containing the text data prepared in previous steps, making it ready for training. The trainer will manage the entire fine-tuning process, including batch management, gradient accumulation, and model updates, ensuring efficient and streamlined training.
**Training Arguments:**

The TrainingArguments class defines essential hyperparameters for fine-tuning, including:

* learning_rate=3e-4: Sets the learning rate for the optimizer, controlling how
quickly the model updates its weights.
* per_device_train_batch_size=1: Specifies a batch size of 1 per device to manage memory constraints, which is essential when working with large models.
* gradient_accumulation_steps=2: Accumulates gradients over 2 steps before updating weights, effectively simulating a larger batch size without requiring additional memory.
* num_train_epochs=1: Sets the number of epochs to 1, meaning the model will pass through the training dataset once.
* fp16=not is_bfloat16_supported() and bf16=is_bfloat16_supported(): Enables mixed precision training based on hardware support, reducing memory usage.
* logging_steps=1: Logs progress at every training step, providing frequent feedback for monitoring training performance.
* optim="adamw_8bit": Utilizes the 8-bit AdamW optimizer, which helps reduce memory usage without compromising optimization quality.
* output_dir="output": Specifies the directory where the fine-tuned model and logs will be saved, ensuring easy access post-training.




```python
import os

# Clear cache
torch.cuda.empty_cache()

# Set environment variable for memory management
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

trainer=SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=training_data,
    dataset_text_field="text",
    max_seq_length=512,
    dataset_num_proc=2,
    packing=True,
    args=TrainingArguments(
        learning_rate=3e-4,
        lr_scheduler_type="linear",
        per_device_train_batch_size= 1,
        gradient_accumulation_steps= 2,
        num_train_epochs=1,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=1,
        optim="adamw_8bit",
        weight_decay=0.01,
        warmup_steps=10,
        output_dir="output",
        seed=0,
    ),
)
```

# Step 5: Fine tune the model


**Training Process:**
Finally, by calling trainer.train(), the training process begins. This method applies all defined hyperparameters to fine-tune the model on the provided dataset. The SFTTrainer manages data batching and gradient accumulation, optimizing resource usage and memory efficiency, allowing the LLAMA3 model to learn from the mental health counseling data.

```python
# Start fine-tuning
trainer.train()
```
